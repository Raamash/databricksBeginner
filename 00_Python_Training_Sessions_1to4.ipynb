### 1.3 Basic String Operations

Strings come with built-in methods — no extra import needed.

````python
city = "  chennai  "

city.strip()        # remove spaces → 'chennai'
city.strip().upper()                # 'CHENNAI'
city.strip().capitalize()           # 'Chennai'
city.strip().replace("chennai", "Mumbai")  # 'Mumbai'

# f-strings — the modern way to embed variables in text
name = "Arjun"
age  = 30
f"{name} is {age} years old"    # 'Arjun is 30 years old'
````

### 2.1 If / Elif / Else

Used to make decisions. Python checks conditions **top to bottom** and stops at the first `True`.

````python
# Classify a sales amount into a tier
sales = 85000

if sales >= 100000:
    tier = "Platinum"
elif sales >= 50000:
    tier = "Gold"
elif sales >= 20000:
    tier = "Silver"
else:
    tier = "Bronze"

tier   # 'Gold'
````

````python
# One-liner (ternary) — useful for simple yes/no
score = 72
result = "Pass" if score >= 60 else "Fail"
result   # 'Pass'
````

### 2.2 For Loop

Iterate over a list, range, or any collection. Think of it as *"do this for each item"*.

````python
departments = ["HR", "Finance", "Engineering", "Sales"]

# Loop and collect results — avoid raw print, use a list
dept_upper = [dept.upper() for dept in departments]
dept_upper   # ['HR', 'FINANCE', 'ENGINEERING', 'SALES']
````

````python
# range(start, stop, step) — useful for numbered iterations
# Generate employee IDs: EMP001, EMP002, ..., EMP005
emp_ids = [f"EMP{str(i).zfill(3)}" for i in range(1, 6)]
emp_ids   # ['EMP001', 'EMP002', 'EMP003', 'EMP004', 'EMP005']
````

````python
# enumerate() gives index + value together
regions = ["North", "South", "East", "West"]

[(idx, region) for idx, region in enumerate(regions, start=1)]
# [(1, 'North'), (2, 'South'), (3, 'East'), (4, 'West')]
````

### 2.3 While Loop

Runs **as long as a condition is True**. Use when you don't know the exact number of iterations in advance.

````python
# Simulate a retry mechanism — common in data pipeline error handling
attempts = 0
max_retries = 3
success = False

while attempts < max_retries and not success:
    attempts += 1
    # Imagine this checks a database connection
    if attempts == 2:      # pretend it succeeds on attempt 2
        success = True

f"Connected after {attempts} attempt(s)"   # 'Connected after 2 attempt(s)'
````

### 2.4 Loop Controls — break & continue

- `break` — exit the loop immediately
- `continue` — skip the current iteration, go to the next

````python
transactions = [500, 200, -999, 800, 100]   # -999 is a sentinel / corrupted marker

valid = []
for amount in transactions:
    if amount == -999:
        break            # stop processing when we hit the bad marker
    if amount < 0:
        continue         # skip negative values
    valid.append(amount)

valid   # [500, 200]
````

### 3.1 List

````python
# A list of monthly sales figures (in thousands)
monthly_sales = [120, 95, 140, 80, 175, 160, 90, 200, 155, 130, 110, 190]

monthly_sales[0]     # first month  → 120
monthly_sales[-1]    # last month   → 190
monthly_sales[3:6]   # slicing: April to June → [80, 175, 160]
max(monthly_sales)   # peak month sales → 200
sum(monthly_sales) / len(monthly_sales)   # average → 137.08...
````

````python
# Common list operations
products = ["Laptop", "Mouse", "Keyboard"]
products.append("Monitor")          # add to end
products.insert(1, "Webcam")        # insert at index 1
products.remove("Mouse")            # remove by value
sorted(products)                    # sorted copy (original unchanged)
````

### 3.2 Tuple

Use tuples for data that **should not change** — coordinates, DB column names, fixed configs.

````python
# Database column definition — should not be modified at runtime
column_schema = ("emp_id", "emp_name", "department", "salary", "hire_date")

column_schema[0]       # 'emp_id'
len(column_schema)     # 5

# Tuple unpacking — very common in Python
lat, lon = (13.0827, 80.2707)   # Chennai coordinates
lat    # 13.0827
````

### 3.3 Set

Use sets when you need **uniqueness** or want to do **set operations** (union, intersection, difference).

````python
# Employees who attended two different trainings
python_attendees = {"Arjun", "Priya", "Karthik", "Deepa", "Arjun"}  # duplicate Arjun
sql_attendees    = {"Priya", "Ravi", "Deepa", "Sanjay"}

len(python_attendees)              # 4 — duplicate removed automatically
python_attendees & sql_attendees   # intersection: attended both → {'Priya', 'Deepa'}
python_attendees | sql_attendees   # union: attended either
python_attendees - sql_attendees   # only Python, not SQL → {'Arjun', 'Karthik'}
````

### 3.4 Dictionary

Use dictionaries for **key-value lookups** — like a row of data where each column has a name.

````python
# A single employee record as a dictionary
employee = {
    "emp_id"    : "E1012",
    "name"      : "Priya Sharma",
    "department": "Finance",
    "salary"    : 72500,
    "is_active" : True
}

employee["name"]                  # 'Priya Sharma'
employee.get("location", "N/A")   # safe access → 'N/A' (key doesn't exist)
list(employee.keys())             # all column names
list(employee.values())           # all values
````

````python
# A list of dictionaries — this is how tabular data looks in Python
employees = [
    {"emp_id": "E1001", "name": "Arjun",  "dept": "Engineering", "salary": 90000},
    {"emp_id": "E1002", "name": "Priya",  "dept": "Finance",     "salary": 72500},
    {"emp_id": "E1003", "name": "Karthik","dept": "Engineering", "salary": 85000},
    {"emp_id": "E1004", "name": "Deepa",  "dept": "HR",          "salary": 60000},
]

# Filter employees from Engineering
[e["name"] for e in employees if e["dept"] == "Engineering"]
# ['Arjun', 'Karthik']
````

### 3.5 String Concatenation & Slicing

### 4.1 List Comprehension

A **concise, Pythonic** way to build a new list from an existing one — replaces a for loop + append.

```
[expression  for item in iterable  if condition]
```

````python
salaries = [72500, 90000, 60000, 85000, 48000, 110000]

# Apply 10% hike to everyone above 70000
hiked = [s * 1.10 if s > 70000 else s for s in salaries]
hiked
# [79750.0, 99000.0, 60000, 93500.0, 48000, 121000.0]
````

````python
# Dictionary comprehension — build a lookup map
names = ["Arjun", "Priya", "Karthik", "Deepa"]
ids   = ["E1001", "E1002", "E1003", "E1004"]

name_to_id = {name: emp_id for name, emp_id in zip(names, ids)}
name_to_id["Priya"]   # 'E1002'
````

### 4.2 Tuple Unpacking

Extract multiple values from a tuple/list in one line — very common when iterating over pairs.

````python
# Swap two variables (no temp variable needed)
a, b = 10, 20
a, b = b, a
a, b   # (20, 10)

# Unpack a row from a database result
row = ("E1001", "Arjun", "Engineering", 90000)
emp_id, name, dept, salary = row

# Extended unpacking — capture the rest with *
first, *middle, last = [10, 20, 30, 40, 50]
first, middle, last   # (10, [20, 30, 40], 50)
````

### 4.3 Try / Except — Error Handling

In data pipelines, bad data is inevitable. `try/except` lets you **handle errors gracefully** instead of crashing.

````python
# Parse salary values from a CSV — some might be missing or malformed
raw_salaries = ["72500", "N/A", "85000", "", "90000"]

def safe_parse(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return None   # return None for bad values instead of crashing

parsed = [safe_parse(v) for v in raw_salaries]
parsed   # [72500.0, None, 85000.0, None, 90000.0]
````

````python
# finally always runs — useful for cleanup (close file, release connection)
def read_config(filename):
    try:
        with open(filename) as f:
            return f.read()
    except FileNotFoundError:
        return "Config file not found — using defaults"
    finally:
        pass   # file is auto-closed by 'with', but finally runs regardless

read_config("missing_config.yaml")   # 'Config file not found — using defaults'
````

### 4.4 Functions

A **function** is a reusable block of code. Good functions do **one thing well** and have a clear name.

````python
# Basic function with default argument
def calculate_bonus(salary, rate=0.10):
    """Return bonus amount based on salary and rate."""
    return round(salary * rate, 2)

calculate_bonus(90000)          # 9000.0  (uses default rate)
calculate_bonus(90000, 0.15)    # 13500.0 (custom rate)
````

````python
# Function returning multiple values (as a tuple)
def salary_stats(salary_list):
    """Return min, max, and average salary."""
    return min(salary_list), max(salary_list), round(sum(salary_list)/len(salary_list), 2)

salaries = [72500, 90000, 60000, 85000, 110000]
low, high, avg = salary_stats(salaries)
low, high, avg   # (60000, 110000, 83500.0)
````

````python
# Lambda — anonymous one-liner function, great for sorting/mapping
employees = [
    {"name": "Arjun",   "salary": 90000},
    {"name": "Priya",   "salary": 72500},
    {"name": "Karthik", "salary": 85000},
]

# Sort employees by salary descending
sorted(employees, key=lambda e: e["salary"], reverse=True)
````

### 4.5 Decorators — A Simple Introduction

A **decorator** wraps a function to add behaviour **without changing the function's code**.
Think of it as: *"Before/after calling this function, also do this extra thing."*

Common real-world uses: logging, timing, authentication checks, retry logic.

````python
import time

# A decorator that measures how long a function takes
def timer(func):
    def wrapper(*args, **kwargs):
        start  = time.time()
        result = func(*args, **kwargs)   # call the original function
        end    = time.time()
        print(f"[timer] {func.__name__} took {end - start:.4f}s")
        return result
    return wrapper

@timer
def process_records(n):
    """Simulate processing n records."""
    return sum(range(n))

process_records(1_000_000)   # prints timing automatically
````

### 4.6 Basic Class & Object Usage

A **class** is a blueprint. An **object** is a specific instance created from that blueprint.

In data engineering, you'll encounter classes in libraries like PySpark (`DataFrame`, `SparkSession`) and SQLAlchemy.

````python
class Employee:
    """Represents a company employee."""

    # __init__ runs when you create a new object
    def __init__(self, emp_id, name, department, salary):
        self.emp_id     = emp_id
        self.name       = name
        self.department = department
        self.salary     = salary

    def apply_raise(self, rate=0.10):
        """Apply a percentage raise and return new salary."""
        self.salary = round(self.salary * (1 + rate), 2)
        return self.salary

    def summary(self):
        return f"{self.name} | {self.department} | ₹{self.salary:,}"


# Create objects from the class
e1 = Employee("E1001", "Arjun",  "Engineering", 90000)
e2 = Employee("E1002", "Priya",  "Finance",     72500)

e1.apply_raise(0.12)   # 100800.0
e1.summary()           # 'Arjun | Engineering | ₹1,00,800.0'
````
